# **Superstore Dataset Preparation to use for Analysis.**

1. Importing pandas to convert file into a pandas Dataframe.

```python
import pandas as pd
```

```python
path_to_csv = '/content/Superstore.csv'
csv = pd.read_csv(path_to_csv, encoding="latin-1")
df = pd.DataFrame(csv)
df
```

2. Removing columns that won't be needed for this analysis by dropping the rows and creating a new dataframe.

```python
df1 = df.drop(columns = ['Unnamed: 0','Row ID', 'Order ID', 'Ship Mode','Customer ID','Customer Name', 'Postal Code', 'Product ID', 'Quantity', 'Discount', 'Ship Date', 'Sub-Category', 'City' ])
df1
```

3. Checking for any null values in dataframe before exporting cleaned dataframe for analysis.

```python
print(pd.isnull(df1).sum())
```

4. Converting database to CSV file to be read and queried in SQL.

```python
df1.to_csv('/content/SuperstoreFeaturesCleaned.csv')
```
